```python
def gradientdescent(x,y):
    m=c=0
    lr=0.001
    n=len(x)
    iterations=179
    for i in range(iterations):
        y_=(m*x)+c
        cost=0
        for f in y-y_:
            
            cost=cost+f**2
        cost=cost/n    
        md=(2/n)*sum((y-y_)*(-x))
        cd=(2/n)*sum(y-y_*-1)
        m=m-lr*md
        c=c-lr*cd
        print(cost)
        
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
gradientdescent(x,y)                
    
```
